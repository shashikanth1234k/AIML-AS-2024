```python
list1 = [1, 2, 3, 4]
list2 = ['a', 'b', 'c', 'd']

import numpy as np

array1 = np.array([5, 6, 7, 8])
array2 = np.array([9, 10, 11, 12])

print(f"Type of list1: {type(list1)}")
print(f"Type of array1: {type(array1)}")

# Multiplication
product = array1 * array2

# Division
quotient = array1 / array2

# Power
power = np.power(array1, 2)

text = "Addition of Two:\n"
combined_text = text + str(array1 + array2)
print(combined_text)

import numpy as np

# Sine
sine_values = np.sin(array1)

# Logarithm (base 10)
log_values = np.log10(array2)

# Logarithm (base 2)
log2_values = np.log2(array2)

# Exponential
exp_values = np.exp(array1)

print(f"Sine values: {sine_values}")
print(f"Logarithm (base 10) values: {log_values}")
print(f"Logarithm (base 2) values: {log2_values}")
print(f"Exponential values: {exp_values}")
```

```python
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40, 45]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

import matplotlib.pyplot as plt

plt.scatter(height, weight)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Scatter Plot: Height vs Weight")
plt.show()


import numpy as np

x = np.arange(len(people))
width = 0.25

plt.bar(x - width, height, width, label="Height")
plt.bar(x, weight, width, label="Weight")

plt.xticks(x, people)
plt.xlabel("People")
plt.ylabel("Values")
plt.title("Bar Chart: Height and Weight")
plt.legend()
plt.show()

plt.hist(age, bins=[20, 25, 30, 35, 40, 45])
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.title("Histogram: Age Distribution")
plt.show()

```
